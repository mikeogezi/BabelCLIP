```python
import json
from PIL import Image
import numpy as np
import ipyplot
from random import sample, seed

_in = json.load(open('ranked/ranked_silver_sample_synsets_l14.json', 'r'))
```

```python
seed(10)
np.random.seed(10)
samples = sample(_in.items(), 50)[25:]

force_b64 = False
start_cnt = 25

for i, (id, v) in enumerate(samples):
    print('{}) {} -> {}\n\t{}'.format(i+1+start_cnt, id, v['gloss'], ', '.join(v['lemmas'])))
    g = v['goldImages']
    m = sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['ex+lemmas_confidence'] or -1, reverse=True)]    
    num_images = min([len(g), len(m), 5])
    if num_images > 1 and not np.array_equiv(g, m):
        r = np.random.rand()
        if r > 0.5:
            first, second = [(g, 'Gold'), (m, 'System')]
        else:
            second, first = [(g, 'Gold'), (m, 'System')]

        for i, img_tup in enumerate([first, second]):
            if i >= 1:
                print('-'*100)
            print(img_tup[1])
            ipyplot.plot_images(img_tup[0], max_images=num_images, img_width=150, show_url=False, labels=range(1, num_images+1), force_b64=force_b64)
    else:
        print('This synset-image mapping was filtered out because both predictions were equivalent')
    
#     print('\nGold')
#     print('----')
#     ipyplot.plot_images(v['goldImages'], max_images=num_images, img_width=128)

#     print('CLIP (lemmas-based)')
#     print('--------------------')
#     sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['lemmas_confidence'], reverse=True)]
#     ipyplot.plot_images(sorted_images, max_images=num_images, img_width=128)
    
#     print('CLIP (gloss-based)')
#     print('-------------------')
#     sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['gloss_confidence'], reverse=True)]
#     ipyplot.plot_images(sorted_images, max_images=num_images, img_width=128)
    
#     print('\nCLIP (example-based)')
#     print('---------------------')
#     sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['example_confidence'] or -1, reverse=True)]
#     ipyplot.plot_images(sorted_images, max_images=num_images, img_width=150)
    
#     print('\nCLIP (ex+gloss-based)')
#     print('---------------------')
#     sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['ex+gloss_confidence'] or -1, reverse=True)]
#     ipyplot.plot_images(sorted_images, max_images=num_images, img_width=150)
    
#     print('\nCLIP (ex+lemmas-based)')
#     print('---------------------')
#     sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['ex+lemmas_confidence'] or -1, reverse=True)]
#     ipyplot.plot_images(sorted_images, max_images=num_images, img_width=150)
```
