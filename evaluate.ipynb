```python
import json
from PIL import Image
import numpy as np
import ipyplot
from random import sample

_in = json.load(open('ranked_silver_sample_synsets_b32.json', 'r'))
sample = sample(_in.items(), 10)
```

```python
for id, v in sample:
    print(id, '->', ', '.join(v['lemmas']))
    
    print('\nGold')
    print('----')
    ipyplot.plot_images(v['goldImages'], max_images=5, img_width=128)

    print('CLIP (lemmas-based)')
    print('--------------------')
    sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['lemmas_confidence'], reverse=True)]
    ipyplot.plot_images(sorted_images, max_images=5, img_width=128)
    
    print('CLIP (gloss-based)')
    print('-------------------')
    sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['gloss_confidence'], reverse=True)]
    ipyplot.plot_images(sorted_images, max_images=5, img_width=128)
    
#     print('\nCLIP (example-based)')
#     print('---------------------')
#     sorted_images = [j['path'] for j in sorted(v['images'], key=lambda x: x['example_confidence'] or -1, reverse=True)]
#     ipyplot.plot_images(sorted_images, max_images=5, img_width=150)
```
